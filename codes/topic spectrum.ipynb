```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("fivethirtyeight")
```

```python
data = pd.read_csv('../data/csv_data/all_corona_data.csv')
data.head()
```

```python
vax_data = data[data['topic'] == 5].copy()
vax_data.head()
```

```python
us_vax = vax_data[vax_data['country'] == 'US']
ca_vax = vax_data[vax_data['country'] == 'CA']
in_vax = vax_data[vax_data['country'] == 'IN']
gb_vax = vax_data[vax_data['country'] == 'GB']
```

```python
fig = plt.figure(figsize=(12, 8))
sns.kdeplot(us_vax['neg_sentim_score'], fill=True, label='US')
sns.kdeplot(ca_vax['neg_sentim_score'], fill=True, label='Canada')
sns.kdeplot(gb_vax['neg_sentim_score'], fill=True, label='UK')
sns.kdeplot(in_vax['neg_sentim_score'], fill=True, label='India')
plt.xticks([0, 0.5, 1], ['Positive', 'Neutral', 'Negative'])
plt.xlabel('Sentiment')
plt.title('Sentiment on vaccination-related articles among countries')
plt.legend()
plt.savefig('../img/vax_country.png', dpi=300)
```

```python
pol_data = data[data['topic'] == 1].copy()
pol_data.shape
```

```python
pol_data['publisher'] = pol_data['publisher'].apply(lambda x: 'Yahoo' if 'yahoo' in x.lower() else x)
pol_data['publisher'].value_counts().head(20)
```

```python
publisher_lst = ['Yahoo', 'Reuters', 'New York Times', 'Business Insider', 'Forbes', 'Associated Press', 
                 'Washington Post', 'New York Post', 'Fox News', 'The Hill']
pol_data_small = pol_data[pol_data['publisher'].isin(publisher_lst)].copy()
```

```python
pol_data_bias = pol_data_small.groupby(['publisher'])['neg_sentim_score'].mean().sort_values().reset_index()
```

```python
pol_data_bias['political_bias'] = [0, -0.5, -1, -2, -1, -1, 0, 2, 0, 1]
```

```python
fig = plt.figure(figsize=(12, 8))
sns.scatterplot(x=pol_data_bias['political_bias'], y=pol_data_bias['neg_sentim_score'], 
                hue=pol_data_bias['political_bias'], palette=sns.color_palette("vlag", as_cmap=True), s=200)
for p, b, n in zip(pol_data_bias['publisher'], pol_data_bias['political_bias'], pol_data_bias['neg_sentim_score']):
    plt.annotate(p, xy=(b, n))
    
plt.yticks([0.4, 0.5, 0.6], ['Positive', 'Neutral', 'Negative'])
plt.xticks([-2, -1, 0, 1, 2], ['Left', 'Lean Left', 'Center', 'Lean Right', 'Right'])
plt.xlabel("Political Bias")
plt.ylabel("Article Sentiment")
plt.legend([])
# plt.savefig("../img/pol_bias.png", dpi=300)
```
